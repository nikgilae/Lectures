```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  #     DataFrame

#  
dob_array = np.array([1990,1991,1992,1993,1994,1995,1996,1997,1998])
y_values = np.arange(len(dob_array))

#  DataFrame  seaborn
df = pd.DataFrame({
    'Year': dob_array,
    'Index': y_values
})

#     
plt.figure(figsize=(10, 4))
sns.regplot(x='Year', y='Index', data=df, ci=None, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel(' ')
plt.ylabel('  ')
plt.title('Scatter-plot   ')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import plotly.express as px
import pandas as pd

#  
dob_array = np.array([1990,1991,1992,1993,1994,1995,1996,1997,1998])
y_values = np.arange(len(dob_array))

#  DataFrame
df = pd.DataFrame({
    'Year': dob_array,
    'Index': y_values
})

#     
fig = px.scatter(
    df, 
    x='Year', 
    y='Index',
    title='Scatter-plot   ',
    trendline='ols',  #  
    trendline_color_override='red'
)

#  
fig.update_layout(
    xaxis_title=' ',
    yaxis_title='  ',
    showlegend=False
)

fig.show()
```
