```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Нужен для работы с DataFrame

# Ваши данные
dob_array = np.array([1990,1991,1992,1993,1994,1995,1996,1997,1998])
y_values = np.arange(len(dob_array))

# Создаем DataFrame для seaborn
df = pd.DataFrame({
    'Year': dob_array,
    'Index': y_values
})

# Построение графика с трендовой линией
plt.figure(figsize=(10, 4))
sns.regplot(x='Year', y='Index', data=df, ci=None, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel('Год рождения')
plt.ylabel('Индекс в массиве')
plt.title('Scatter-plot с трендовой линией')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import plotly.express as px
import pandas as pd

# Ваши данные
dob_array = np.array([1990,1991,1992,1993,1994,1995,1996,1997,1998])
y_values = np.arange(len(dob_array))

# Создаем DataFrame
df = pd.DataFrame({
    'Year': dob_array,
    'Index': y_values
})

# Построение графика с трендовой линией
fig = px.scatter(
    df, 
    x='Year', 
    y='Index',
    title='Scatter-plot с трендовой линией',
    trendline='ols',  # Линейная регрессия
    trendline_color_override='red'
)

# Настройка оформления
fig.update_layout(
    xaxis_title='Год рождения',
    yaxis_title='Индекс в массиве',
    showlegend=False
)

fig.show()
```
