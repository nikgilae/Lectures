```python
class MixinSpeakable:
    def speak(self):
        print(f"{self.name} сказал {self.says}")

class MixinAnimated:
    is_animated = True
    def animete(self):
        print(f"Анимация {self.name}")

class MixinFunny:
    def make_laugh(self):
        print(f"{self.name} пошутил")

class MixinCollectable:
    is_collectable = True
    def collect(self):
        print(f"{self.name} можно коллекционировать")

class MixinActionable:
    def action(self):
        print(f"{self.name} делает действие {self.action_character}")

class MixinPossible:
    def pose(self):
        print(f"{self.name} принял позу")

class MixinCostumeWearable:
    def wearing_costume(self):
        print(f"На {self.name} можно одет костюм {self.costume}")
        

class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    def __init__(self,name):
        self.name = name

class BaseHuman:
    is_human = True

class BaseInActionCharacter(BaseCharacter, MixinActionable):
    action_character = True

class BaseFunkoPop(BaseCharacter,MixinCollectable):
    def display(self):
        print(f"{self.name} не распечатаная в коробке")

class BaseCosplayer(BaseCharacter,BaseHuman,MixinPossible,MixinCostumeWearable):
    costume = "стандартный"

class Shrek(BaseCharacter):
    def __init__(self):
        self.name = "Shrek"
        self.says = "Я зеленый и страшный"

class PussinBoots(BaseCharacter):
    def __init__(self):
        self.name = "PussinBoots"
        self.says = "Я могу победить вас всех взглядом"

class Donkey(BaseCharacter):
    def __init__(self):
        self.name = "Donkey"
        self.says = "Я люблю женщин с формами"
        
class JackHorner(BaseCharacter):
    def __init__(self):
        self.name = "JackHorner"
        self.says = "Я Джек и я заберу у тебя все !"

class ShrekInAction(BaseInActionCharacter,Shrek):
    action_character = "Надул жабу через попу"
    
class PussinBootsInAction(BaseInActionCharacter,PussinBoots):
    action_character = "Вылизывается"

class DonkeyInAction(BaseInActionCharacter,Donkey):
    action_character = "Много разговаривает "

class JackHornerInAction(BaseInActionCharacter,JackHorner):
    action_character = "Крадет магические предметы, существ и людей"
    
class ShrekFunkoPop(BaseFunkoPop):
    def __init__(self):
        self.name = "Shrek"

class PussinBootsFunkoPop(BaseFunkoPop):
    def __init__(self):
        self.name = "PussiInBoots"

class DonkeyFunkoPop(BaseFunkoPop):
    def __init__(self):
        self.name = "Donkey"

class JackHornerFunkoPop(BaseFunkoPop, JackHorner):
    def __init__(self):
        self.name = "JackHorner"

class ShrekCosplayer(BaseCosplayer, Shrek):
    costume = "Шрека"

class DonkeyCosplayer(BaseCosplayer, Donkey):
    costume = "Осла"

class PussInBootsCosplayer(BaseCosplayer, PussinBoots):
    costume = "Кота в сапогах"

class JackHornerCosplayer(BaseCosplayer, JackHorner):
    costume = "Джека Хорнера"    

if __name__ == "__main__":
    shrek = Shrek()
    pussin_boots = PussinBoots()
    donkey = Donkey()
    jack_horner = JackHorner()

    shrek_action = ShrekInAction()
    pussin_boots_action = PussinBootsInAction()
    donkey_action = DonkeyInAction()
    jack_horner_action = JackHornerInAction()

    shrek_pop = ShrekFunkoPop()
    pussin_boots_pop = PussinBootsFunkoPop()
    donkey_pop = DonkeyFunkoPop()
    jack_horner_pop = JackHornerFunkoPop()

    shrek_cos = ShrekCosplayer()
    donkey_cos = DonkeyCosplayer()
    pussinboots_cos = PussInBootsCosplayer()
    jackhorner_cos = JackHornerCosplayer()

    for char in [shrek, pussin_boots, donkey, jack_horner]:
        char.speak()
        char.make_laugh()
    print()

    for char in [shrek_action, pussin_boots_action, donkey_action, jack_horner_action]:
        char.action()
    print()
    
    for char in [shrek_pop, pussin_boots_pop, donkey_pop, jack_horner_pop]:
        char.display()
    print()
    
    for char in [shrek_cos,donkey_cos,pussinboots_cos,jackhorner_cos]:
        char.speak()
        char.make_laugh()
        char.pose()
        char.wearing_costume()
        print("Костюм:", char.costume)
        print()
```
